# Phase 2 — Real Data Integration Test

This notebook allows you to test the full Aegis Finance LangGraph orchestrator using your local database and local LLM.

```python
import nest_asyncio
from aegis.graph import process_query

# Required for running async in Jupyter
nest_asyncio.apply()
```

## 1. Test SQL Flow (Personal Financial)
This should query your `v_*` views.

```python
query = "What is my current total net worth?"
result = await process_query(query)

print(f"Final Answer: {result.get('final_answer')}")
print("\nSQL Results:")
for row in result.get("sql_result", []):
    print(row)
```

## 2. Test RAG Flow (Market Knowledge)
This should query your `pgvector` knowledge base.

```python
query = "What is a CEDEAR and how does the conversion ratio work?"
result = await process_query(query)

print(f"Final Answer: {result.get('final_answer')}")
print("\nRetrieved Chunks:")
for chunk in result.get("rag_chunks", []):
    print(f"- Source: {chunk['source_title']} | Content: {chunk['content'][:100]}...")
```

## 3. Test Privacy Middleware
Check how PII is redacted.

```python
query = "My CUIT is 20-12345678-9. What can you tell me about my taxes?"
result = await process_query(query)

privacy = result.get("privacy_output", {})
print(f"Sanitized Query: {privacy.get('sanitized_query')}")
print(f"Risk Score: {privacy.get('risk_score')}")
print(f"Redaction Map: {privacy.get('redaction_map')}")
```
